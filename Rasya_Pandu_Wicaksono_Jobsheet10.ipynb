Praktikum 01 Bubble Sort

```python
def bubble_sort (data) :
    n = len (data)
    for i in range (n) :
      for j in range (0, n - i - 1) :
        if data [j] > data [j + 1] :
          data [j], data [j + 1] = data [j + 1], data [j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print ("Data sebelum diurutkan:", angka)
print ("Data setelah Bubble Sort:", bubble_sort(angka))
```

Praktikum 02 Selection Sort

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03 Insertion Sort

```python
def insertion_sort(data) :
  n = len(data)
  for i in range(1, n) :
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 04 Studi Kasus: Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data) :
  n = len(data)
  for i in range(n) :
    for j in range(0, n - i - 1) :
      if data[j][1] < data[j + 1] [1] :
        data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [ ("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print ("Data sebelum diurutkan:", mahasiswa)
print ("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

Praktikum 06 Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False) :
  n = len(data)
  for i in range(n) :
    for j in range(0, n - i - 1) :
      if reverse:
        if data[j][key_index] < data[j +1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
        else:
          if data[j][key_index] > data[j + 1][key_index]:
              data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [ ("Buku", 15000),  ("Pensil", 5000), ("Tas", 75000), ("Pulpen",10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(),1, reverse=True))
```

Penugasan 1 Eksperimen Bubble Sort

```python
def bubble_sort_descending(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", data)

sorted_data = bubble_sort_descending(data)
print("Data setelah diurutkan (Descending):", sorted_data)
```

Penugasan 2 Eksperimen Selection Sort

```python
def selection_sort_max_to_end(arr):
    n = len(arr)
    for i in range(n - 1, 0, -1):
        max_idx = 0
        for j in range(1, i + 1):
            if arr[j] > arr[max_idx]:
                max_idx = j
        arr[max_idx], arr[i] = arr[i], arr[max_idx]

    return arr

data = [29, 10, 14, 37, 13]
print("Data asli:", data)
print("Data terurut:", selection_sort_max_to_end(data))
```

Penugasan 3 Eksperimen Insertion Sort

```python
def insertion_sort_mahasiswa(arr):
    for i in range(1, len(arr)):
        key_item = arr[i]
        j = i - 1
        while j >= 0 and arr[j].lower() > key_item.lower():
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

    return arr

daftar_mahasiswa = ["Zaki", "Andi", "Budi", "Candra", "Dewi", "Eka"]
print("Sebelum diurutkan:", daftar_mahasiswa)

sorted_nama = insertion_sort_mahasiswa(daftar_mahasiswa)
print("Setelah diurutkan alfabetis:", sorted_nama)
```

Penugasan 5 Analisis Kompleksitas

```python
import random

def bubble_sort_count(arr):
    n = len(arr)
    count = 0
    temp_arr = arr.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if temp_arr[j] > temp_arr[j + 1]:
                temp_arr[j], temp_arr[j + 1] = temp_arr[j + 1], temp_arr[j]
    return count

def selection_sort_count(arr):
    n = len(arr)
    count = 0
    temp_arr = arr.copy()
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            count += 1
            if temp_arr[j] < temp_arr[min_idx]:
                min_idx = j
        temp_arr[i], temp_arr[min_idx] = temp_arr[min_idx], temp_arr[i]
    return count

def insertion_sort_count(arr):
    n = len(arr)
    count = 0
    temp_arr = arr.copy()
    for i in range(1, n):
        key = temp_arr[i]
        j = i - 1
        while j >= 0:
            count += 1
            if temp_arr[j] > key:
                temp_arr[j + 1] = temp_arr[j]
                j -= 1
            else:
                break
        temp_arr[j + 1] = key
    return count
sizes = [10, 50, 100]

print(f"{'Ukuran':<10} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 50)

for n in sizes:
    data_acak = [random.randint(1, 1000) for _ in range(n)]
    b_count = bubble_sort_count(data_acak)
    s_count = selection_sort_count(data_acak)
    i_count = insertion_sort_count(data_acak)

    print(f"{n:<10} | {b_count:<10} | {s_count:<10} | {i_count:<10}")
```

Penugasan 6 Proyek Mini

```python
%%writefile manajemen_barang.py

def urutkan_barang(daftar, kriteria, arah):
    n = len(daftar)
    for i in range(1, n):
        key_item = daftar[i]
        j = i - 1
        while j >= 0:
            harus_tukar = False
            val_j = daftar[j][kriteria].lower() if kriteria == 'nama' else daftar[j][kriteria]
            val_key = key_item[kriteria].lower() if kriteria == 'nama' else key_item[kriteria]

            if arah == 'asc':
                if val_j > val_key:
                    harus_tukar = True
            else:
                if val_j < val_key:
                    harus_tukar = True
            if harus_tukar:
                daftar[j + 1] = daftar[j]
                j -= 1
            else:
                break
        daftar[j + 1] = key_item
    return daftar
```

```python
import manajemen_barang

def main():
    list_barang = []
    print("--- Input Daftar Barang ---")
    n = int(input("Masukkan jumlah barang: "))
    for i in range(n):
        print(f"\nBarang ke-{i+1}")
        nama = input("Nama barang: ")
        harga = int(input("Harga barang: "))
        list_barang.append({'nama': nama, 'harga': harga})
    print("\n--- Menu Urutkan ---")
    print("1. Berdasarkan Nama")
    print("2. Berdasarkan Harga")
    pilih_kriteria = input("Pilih (1/2): ")
    kriteria = 'nama' if pilih_kriteria == '1' else 'harga'

    print("\n--- Pilih Arah ---")
    print("1. Ascending (Kecil ke Besar / A-Z)")
    print("2. Descending (Besar ke Kecil / Z-A)")
    pilih_arah = input("Pilih (1/2): ")
    arah = 'asc' if pilih_arah == '1' else 'desc'

    hasil = manajemen_barang.urutkan_barang(list_barang, kriteria, arah)

    print(f"\n--- Hasil Pengurutan ({kriteria} - {arah}) ---")
    print(f"{'No':<3} | {'Nama Barang':<15} | {'Harga':<10}")
    print("-" * 35)
    for idx, b in enumerate(hasil):
        print(f"{idx+1:<3} | {b['nama']:<15} | Rp{b['harga']:,}")
if __name__ == "__main__":
    main()
```
